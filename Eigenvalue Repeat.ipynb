# Eigenvalue Repeat
What happens if one keeps on finding the eigenvalues of the matrix made of eigenvectors?

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = np.array([[1,0.4],[0.4,1]])
A = np.array([[1,0.4,0],[0.4,2,0.4],[0,0.4,3]])


n_iter = 10000
u_vec = []
for i in range(n_iter): 
    u,v = np.linalg.eig(A) 
    u_vec.append(u)
    A = v.T
u_vec = np.array(u_vec)
```

```python
plt.plot(u_vec[:,0].real)
plt.plot(u_vec[:,0].imag)
```

```python
plt.plot(u_vec[:,1].real)
plt.plot(u_vec[:,1].imag)
```

```python
plt.plot(np.arctan2(u_vec[:,0].imag,u_vec[:,0].real))
```

```python
plt.plot(np.arctan2(u_vec[:,1].imag,u_vec[:,1].real))
```

```python
angles = np.arctan2(u_vec[:,0].imag,u_vec[:,0].real) 

f,bins = np.histogram(angles,bins=100)

plt.plot(bins[:-1],f)
```

# Notes
- The eigenvalues seem to get randomly distributed on the unit circle
- An interesting question is, what is the set of matrices for which this transformation leads to fixed points?
- There is a non-trivial distribution of the phase, when the eigenvalues get distributed on the unit circle.
