# Bose-Einstein Condensation

Consider the function as $f(z,x)$ defined:
   
$$ f(z,x) = \frac{z x^{1/2}}{\exp(x) - z} $$
   
and the integral over $x$,

$$I(z) = \int_0^{\infty} f(z,x) dx $$

Here, $z$ is a positive real number.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.integrate
```

## Spectral Number density function $f$

```python
def f(z,x):
    return z*np.sqrt(x)/(np.exp(x) - z)
```

```python
x_vec = np.linspace(0,2,1000)

plt.figure(dpi=150)
colors = plt.cm.viridis(np.linspace(0,1,6))
for i,z in enumerate([0.1,0.5,0.9,0.99]):
    f_vec = f(z,x_vec)
    plt.plot(x_vec,f_vec,color=colors[i],label="z = {:.1f}".format(z))
plt.legend()
plt.xlabel(r"$x$",fontsize=14)
plt.ylabel(r"$f(z,x)$",fontsize=14)
ax = plt.gca()
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.yaxis.set_major_locator(plt.MaxNLocator(5))
```

Observe in the above plot, as $z \rightarrow 1$ i.e.$\mu \rightarrow 0$ ($z = e^{\beta \mu}$), the $x = 0$ looks like it is getting increasingly occupied. This is the formation of Bose-Einstein condensate.

## Total number density function $I$

```python
def I(z):
    return scipy.integrate.quad(lambda x : f(z,x),0,np.inf)[0]
```

```python
z_vec = np.linspace(0,1,1000)
plt.figure(dpi=150)

I_vec = np.array([I(z) for z in z_vec])
plt.plot(z_vec,I_vec,color="r")
plt.xlabel(r"$z$",fontsize=14)
plt.ylabel(r"$I(z)$",fontsize=14)
ax = plt.gca()
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.yaxis.set_major_locator(plt.MaxNLocator(5))
```
